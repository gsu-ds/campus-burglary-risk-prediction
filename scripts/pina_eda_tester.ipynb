```python
import pandas as pd
from rich.console import Console
from rich.table import Table
from rich.panel import Panel

console = Console()

try:
    df = pd.read_csv("/workspaces/campus-burglary-risk-prediction/data/november_eda/harini_cleaned/CLEANED_GSU_Burglary_Larceny_1mi.csv", nrows=0) 

    col_names = df.columns.tolist()

    console.print("\n[bold cyan]Columns as a Table:[/bold cyan]")
    table = Table(title="GSU-Larceny-1mi-ColNames")
    table.add_column("Index", style="dim", width=6, justify="right")
    table.add_column("Column Name", style="bold")

    for i, name in enumerate(col_names):
        table.add_row(str(i), name)
    
    console.print(table)

except FileNotFoundError:
    console.print(f"Error: The file was not found.", style="bold red")
except Exception as e:
    console.print(f"An error occurred: {e}", style="bold red")
```
