##### Attempt to use geopandas to reverse geocode neighborhoods to fill empty values by their long/lat.

```python
import geopandas as gpd
from shapely.geometry import Point
from pathlib import Path

PATH = "/workspaces/campus-burglary-risk-prediction/data/november_eda/pina_cleaned/cleaned_full_atl_v3.csv"

df = pd.read_csv(PATH)

row_num = df.shape[0]
col_num = df.shape[1]
col_names = df.columns.tolist()

print('Num of Cols in DF:', col_num,
      '\nNum of Rows in DF:', row_num,
      df.shape,
      'Col Names:', col_names)
```

```python
SF_PATH = '/workspaces/campus-burglary-risk-prediction/data/november_eda/pina_cleaned/atl_neighborhood/Official_Neighborhoods_-_Open_Data.shp'
neighborhoods = gpd.read_file(SF_PATH)

print("[bold green]CSV Shape Confirmed: [/bold green]",df.shape)
print("[bold green]Shape File read.[/bold green]")
print(neighborhoods.head())
```

```python
gdf_points = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df.Longitude, df.Latitude),
    crs="EPSG:4326"
)

invalid_points = gdf_points.geometry.is_empty | gdf_points.geometry.isna()
print(f"Found {invalid_points.sum()} invalid or empty points.")


#print(gdf_points.head())
print("[bold green]CSV converted to GeoDataFrame using standard CRS for lat/lon.[/bold green]")

gdf_points = gdf_points.to_crs(neighborhoods.crs)
print("[bold green]CRS aligned.[/bold green]")
```

```python
joined_gdf = gpd.sjoin(gdf_points, neighborhoods, how="left", predicate="within")

joined_gdf = joined_gdf.drop(columns='index_right', errors='ignore')
print("[bold green]Spatial join complete.[/bold green]")
```

```python
import matplotlib.pyplot as plt

# 1. Isolate your missing points
missing_points_df = joined_gdf[joined_gdf['NhoodName'].isna()]
print(f"Plotting {len(missing_points_df)} missing points.")

# 2. Create a plot
fig, ax = plt.subplots(figsize=(8, 8))

# 3. Plot the neighborhood map as the base layer
neighborhoods.plot(ax=ax, color='lightgray', edgecolor='white')

# 4. Plot your missing points on top
missing_points_df.plot(ax=ax, color='red', markersize=5, label='Missing Points')

ax.legend()
plt.title("Map of Points That Did NOT Match a Neighborhood")
plt.show()
```

- The red dots on the bottom represent cities outside of Atlanta. East Point, Hapeville, and College Park all had data in the spreadsheet. 
    - This could not be mapped to neighborhoods in Atlanta, because they are not actually in Atlanta. v5 will mark these points as such. 
- There is a hole where Emory is, which means this map must be pre 2018.
